# Code for Analyzing Songs Recognized by Shazam

```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
song_info = f'{cwd}/../test-shazam-api/songs_info.csv'
no_song_found = f'{cwd}/../test-shazam-api/no_match_songs.txt'

# videos with no song recognized
with open(no_song_found, 'r') as file:
    # Read the entire content of the file into a string
    content = file.read()

id_no_match = content.split('\n')
print("Number of videos without a song: ", len(id_no_match))

song_info = pd.read_csv(song_info)
print("Number of videos with a song: ", len(song_info))
song_info
#no_song = pd.read
```

### Now, try to find most popular songs and artists

```python
from collections import Counter

artists = Counter(song_info['artist']).most_common(15)
print('Most common artists: ')
for value, count in artists:
    print(f"{value}: {count}")

songs = Counter(song_info['track_name']).most_common(15)
print('\nMost common songs: ')
for value, count in songs:
    print(f"{value}: {count}")
```

```python
artists
singers = []
for a, b in artists: 
    singers.append(a)
```

```python
singers
```

```python
songs
tracks = []
for a, b in songs:
    tracks.append(a)
```

```python
tracks
```

```python
viral = pd.DataFrame(columns=['artists', 'songs'])
for index, row in song_info.iterrows():
    if row.at['artist'] in singers or row.at['track_name'] in tracks:
        if not viral['songs'].isin([row.at['track_name']]).any():
            viral.loc[len(viral)] = [row.at['artist'], row.at['track_name']]
```
